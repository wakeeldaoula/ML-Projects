```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('J:\\4-Fourth Smester\\coding_semester4\\AI Lab\\ML\\4-KNN\\customer_purchase_data.csv')
```

```python
data.head()
```

```python
feature_cols = ['NumberOfPurchases', 'AnnualIncome']
X = data[feature_cols]
```

```python
X.head()
```

```python
Y = data[['TimeSpentOnWebsite']]
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.1, random_state=1) # if we don't use test_size here then it will test on 25% of data by default
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
model.intercept_ # this will show the intercept (b) in linear regression   y = mx + b
model.coef_ # this will show the coefficient (slope m) in linear regression    y = mx + b    
```

```python
zip(feature_cols, model.coef_)
```

Evaluation Functions
-> Mean Squared Error (MSE)
-> Root mean squared error (RMSE)
-> Mean Absolute Error (MAE) this is just average error
-> R^2   (R squared)

```python
from sklearn.metrics import mean_squared_error as mse, root_mean_squared_error as rmse, mean_absolute_error as mae, r2_score as rs
```

```python
print(mse(Y_pred, Y_test)) # huge error, bcz this dataset is not good for linear regression and we are just studying examples
print(rmse(Y_pred, Y_test))
print(mae(Y_pred, Y_test))
print(rs(Y_pred, Y_test))
```
