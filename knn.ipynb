```python
'''
A retail company wants to predict customer purchasing behavior based on their age, salary, and past purchase
history. The company aims to use k-nearest neighbors (knn) algorithm to classify customers into potential buying
groups to personalize marketeing strategies. This predictive model will help the company 
understand and target specific customer segments more effectively, thereby increasing sales and customer satisfaction
'''
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # he StandardScaler from scikit-learn is used to adjust your data so that:
# Mean is 0: It makes the average of your data points zero.
# Standard Deviation is 1: It scales the data so that the spread (how much the values vary) is normal.
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Designing some source of data
data = np.array([[25,10000,1],[30,50000,3],[35,60000,5],[50,100000,5]]) # age, sallary, number of products purchased
data
```

```python
labels = np.array([0,1,2,2]) # 0: low, 1: mediocre, 2: high 
# we have to identify high level customers
# lebels for customers given above
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data,labels,test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler() # defining object for StandardScaler to usee for normalizing data
X_train = scaler.fit_transform(X_train) 
X_test = scaler.fit_transform(X_test) 
```

```python
knn = KNeighborsClassifier(n_neighbors=2) # defining the object fo KNeighborsClassifier
knn.fit(X_train,Y_train)
```

```python
accuracy = knn.score(X_test,Y_test)
print("Accuracy:",accuracy * 100)
```
