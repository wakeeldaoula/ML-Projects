```python
'''
A company is looking to develop a text classification model to categorize customer reviews into positive or negative
sentiments. They want to use Naive Bayes classification to automatically analyze and classify the reviews they receive,
aiming to understand customer satisfaction levels and sentiments. The company desires a model that can accurately predict
whether a customer review expresses a positive or negative sentiment.
'''
```

```python
# step 1
import numpy as np
import pandas as pd
# from sklearn.model_selection import train_test_split  We don't need to import it because CountVectorizer can fit and fit_transform methods
from sklearn.feature_extraction.text import CountVectorizer # for tokenizing the strings
from sklearn.naive_bayes import MultinomialNB # for naive bayes 
# step 2
reviews = ["The product is very good",
            "The product works correctly and most satisfying", "waste of money",
              "waste of time",
           "I love this product", "Its amazing", 
           "First time I am thinking that 5 stars are not enough to judge this product"]
sentiments = np.array([1,1,0,0,1,1,1]) # 0 is for negative review and 1 is for positive review
```

```python
# step 3
vectorizer = CountVectorizer() # defining an object
X = vectorizer.fit_transform(reviews) # it gives shape and number of tokens in all strings in the review list
```

```python
classifier = MultinomialNB()
classifier.fit(X,sentiments)
```

```python
accuracy = classifier.score(X,sentiments)
print("Accuracy:",accuracy*100,"%")
```

```python
# manual predictions
text1 = input("Enter some text:")
text2 = input("Enter some text:")
text3 = input("Enter some text:")
text4 = input("Enter some text:")
text5 = input("Enter some text:")
input_feature = [[text1,text2,text3,text4,text5]]
vectorized_input = vectorizer.fit_transform(input_feature)
classifier.predict(vectorized_input)
```
